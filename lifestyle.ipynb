## Quick quiz
Try making a stackplot with your own weekly schedule!

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]

# amount of time i spend on my daily activities in hrs (in summer holidays)
# these are the major activities i do other than bathing, eating meals and other things
sleeping=[8,8.5,7.5,9,9,7,8]
gym=[1.5,2,0,1.55,2.2,1.5,0]
video_games=[1, 49/60, 1, 30/60, 0, 39/60, 25/60]
programming=[1, 45/60, 30/60, 20/60, 1, 35/60, 1]
show=[30/60, 10/60, 5/60, 1, 45/60, 2, 1]

# visualising and anaylising the data of my schedule over 1 week
label=["Sleep", "Gym", "Gaming", "Programming", "Show"]
colors=["lightgreen", "red", "cyan", "blue", "grey"]
plt.stackplot(days, sleeping, gym, video_games, programming, show, labels=label, colors=colors)
plt.xlabel("Days")
plt.ylabel("Total hours spent on major activities")
plt.legend()
plt.tight_layout()
plt.show()
```
